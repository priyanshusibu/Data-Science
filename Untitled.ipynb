```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-10, 10, 400)

# Calculate the y values for each function
y1 = x * np.sin(x)
y2 = x ** 2
y3 = np.sin(x)
y4 = x ** 3 * np.sin(x)

# Plotting the functions
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r'$x\sin(x)$')
plt.plot(x, y2, label=r'$x^2$')
plt.plot(x, y3, label=r'$\sin(x)$')
plt.plot(x, y4, label=r'$x^3\sin(x)$')

# Adding labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of Functions')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
