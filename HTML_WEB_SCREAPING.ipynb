```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
dataset=pd.read_csv("Mall_Customers.csv")
df=dataset
x_mean=np.mean(df['Age'])
y_mean=np.mean(df['Annual_Income_(k$)'])
print("X mean is :",x_mean)
print("Y mean is :",y_mean)
xy_sum=0
for i,j in zip(df['Age'],df['Annual_Income_(k$)']):
    xy_sum=xy_sum+(i*j)
print("Summetion of xy is :",xy_sum)
x_squre_sum=0
for i in df['Age']:
    x_squre_sum=x_squre_sum+(i**2)
print("X^2 sum is:",x_squre_sum)
n=len(df)
slope=(xy_sum+(n*x_mean*y_mean))/(x_squre_sum-(n*(x_mean**2)))
intercept=y_mean-(slope*x_mean)
print("Intercept calculation:",intercept)
#formula y=slope*x  + intercept
y_axis=[]
for i in df['Age']:
    temp=(slope*i)+intercept
    y_axis.append(temp)
plt.figure(figsize=(10,10))
plt.plot(df['Age'], y_axis)
plt.scatter(df['Age'], df['Annual_Income_(k$)'], linestyle='-', color='red')
plt.xlabel("Age")
plt.ylabel("Annual Income")
plt.show()
```
