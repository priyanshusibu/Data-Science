```python
import matplotlib.pyplot as plt

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
max_temperatures = [17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18]
min_temperatures = [-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58]
plt.figure(figsize=(10, 6))
plt.plot(months, max_temperatures, marker='o', color='r', label='Max Temperature (C)')
plt.plot(months, min_temperatures, marker='o', color='b', label='Min Temperature (C)')
plt.title('Temperature Extremes in a Region of India')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

male_age = [53, 51, 71, 31, 33, 39, 52, 27, 54, 30, 64, 26, 21, 54, 52, 20, 59, 32]
female_age = [53, 65, 68, 21, 75, 46, 24, 63, 61, 24, 49, 41, 39, 40, 25, 54, 42, 32, 48, 23, 23]

# Calculate the mean age for males and females
mean_male_age = sum(male_age) / len(male_age)
mean_female_age = sum(female_age) / len(female_age)

# Plotting the bar graph
labels = ['Male', 'Female']
mean_ages = [mean_male_age, mean_female_age]

plt.bar(labels, mean_ages, color=['blue', 'pink'])
plt.title('Mean Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Age')
plt.show()
```
