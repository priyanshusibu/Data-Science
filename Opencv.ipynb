```python
hsv_image=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
```

```python
hsv_image
```

```python
plt.imshow(hsv_image)
```

```python
hue,sat,val=cv2.split(hsv_image)
```

```python
val
#It will represent the value for ever singe pixel
```

```python
 val+50
```

```python
add_val=50 #Here we want the increase the brightness by 50
val+add_val
```

```python
import numpy as np
```

```python
#WE ARE ONLY CHANGINFTHE BRIGHTNESS SO WE CAN DO THAT BY THE HELP OF VLAUE ATTRIBUTE OF (hsv) FORMAT
```

```python
new_val=val.astype(np.int64)+add_val
new_val[new_val>255]=255
new_val_uint8=new_val.astype(np.uint8)
```

```python
new_image_hsv=cv2.merge([hue,sat,new_val_uint8])
```

```python
new_image_bgr=cv2.cvtColor(new_image_hsv,cv2.COLOR_HSV2BGR)
```

```python
cv2.imshow('img',new_image_bgr) #Basically create the window and put the image in
cv2.waitKey(0) #Wait for the input like close button or minimize button like that
cv2.destroyAllWindows() #Once everything is done then distory all the window
```
